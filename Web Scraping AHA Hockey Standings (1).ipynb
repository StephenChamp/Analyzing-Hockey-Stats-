# Web Scraping AHA Hockey Standings

Prior to moving to Dallas, Texas I played in a hockey league called the Adult Hockey Association. They keep track of player and team stats and I always thought it would be fun to scrape the information from their website to do my own analytics on the data. Here I will be doing some basic scraping of the stats around the league I play in. I am currently on the Battle Cats and we are having a pretty good season so far.

### Import packages needed to Web Scrape, I will be using Beautiful Soup

```python
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
```

### We choose the website we want to look at and specifically the current league I play in

```python
website ="https://www.ahahockey.com/standings/twin-cities?selected_tier=316"
```

### Download the HTML code and start a BeautifulSoup Object

```python
response = urllib.request.urlopen(website)
the_website = response.read()
soup = BeautifulSoup(the_website, "html.parser")
```

### We can find the table with the data using the following

```python
for x in soup.find_all('td'):
    print(x.get_text().strip())
```

### Define some lists to store the data in and iterate over the list to get all of the data

```python
name = []
gp = []
W = []
L = []
T = []
OTL = []
Pts = []
SF = []
SA = []
GF = []
GA = []
PIMS = []
Seed = []
counter = 1
for x in soup.find_all('td'):
    if counter == 1:
        name.append(x.get_text().strip())
        counter += 1
    elif counter == 2:
        gp.append(x.get_text().strip())
        counter += 1
    elif counter == 3:
        W.append(x.get_text().strip())
        counter += 1
    elif counter == 4:
        L.append(x.get_text().strip())
        counter += 1
    elif counter == 5:
        T.append(x.get_text().strip())
        counter += 1
    elif counter == 6:
        OTL.append(x.get_text().strip())
        counter += 1
    elif counter == 7:
        Pts.append(x.get_text().strip())
        counter += 1
    elif counter == 8:
        SF.append(x.get_text().strip())
        counter += 1
    elif counter == 9:
        SA.append(x.get_text().strip())
        counter += 1
    elif counter == 10:
        GF.append(x.get_text().strip())
        counter += 1
    elif counter == 11:
        GA.append(x.get_text().strip())
        counter += 1
    elif counter == 12:
        PIMS.append(x.get_text().strip())
        counter += 1
    elif counter == 13:
        Seed.append(x.get_text().strip())
        counter = 1
```

### Turn our Lists into a pandas DataFrame

```python
Hockey_data = pd.DataFrame([name,gp, W, L, T, OTL, Pts, SF, SA, GF, GA, PIMS, Seed])
Hockey_data
```

### Lets Make the Dataframe match the website

```python
Hockey_data = Hockey_data.transpose()
titles = Hockey_data.iloc[0]
Hockey_data = Hockey_data[1:]
Hockey_data.columns = titles
Hockey_data
```
